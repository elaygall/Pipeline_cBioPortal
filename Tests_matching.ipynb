```python
genes = np.array(df2['Hugo_Symbol'])
genes
```

```python
print(len(genes))
```

```python
genes2 = np.array(df['GENE'])
```

```python
print(len(genes2))
```

```python
Inter = np.intersect1d(genes, genes2)
```

```python
print(Inter) #genes found in both test
```

```python
ODC_unic = len(genes2) - len(Inter)
MutSig_unic = len(genes) - len(Inter)
Interception = len(Inter)
```

```python
import matplotlib.pyplot as plt

graf = pd.DataFrame({'ALGO': ['OncodriveCLUST', 'MutSig', 'Interception'], 'Number_of_genes': [ODC_unic, MutSig_unic, Interception]})
graf.Number_of_genes.groupby(graf.ALGO).sum().plot(kind='pie')
plt.axis('equal')
plt.show()
```
